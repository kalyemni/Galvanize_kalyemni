```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.close('all')
```

```python
df = pd.read_csv("./worldcup-2018.csv")
```

```python
df.groupby(['Group']).mean()
```

```python
Team_Ranking = df[['Team', 'Current \nFIFA rank']]
Matches = df[['Team', 'First match \nagainst']]
```

```python
Team_Ranking 
```

```python
plt.figure()
```

```python
Team_Ranking.plot(kind='bar');
```

```python
a= Team_Ranking.iloc[::2]
First_Matches = Matches.iloc[::2]
b= Team_Ranking.iloc[1::2]
fram = [a,b]
pd.concat(fram)
```

```python
len(list(First_Matches['Team'] +' vs. '+ First_Matches['First match \nagainst']))
```

```python
len(list(a['Current \nFIFA rank']))
len(list(b['Current \nFIFA rank']))
len(['M1', 'M2', 'M3', 'M4', 'M5', 'M6', 'M7', 'M8', ' 'M9', 'M10', 'M11', 'M12', 'M13', 'M14', 'M15', 'M16'])
```

```python

labels = list(First_Matches['Team'] +' vs. '+ First_Matches['First match \nagainst'])
men_means = list(a['Current \nFIFA rank'])
women_means = list(b['Current \nFIFA rank'])

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='Team1')
rects2 = ax.bar(x + width/2, women_means, width, label='Team2')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Team Ranks By first match')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(5, 10),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```

```python
# Python dictionary

import matplotlib.pyplot as plot
inflationAndGrowth  = {"Team 1": list(a['Current \nFIFA rank']),
                       "Team 2": list(b['Current \nFIFA rank'])};

index               = list(First_Matches['Team'] +' vs. '+ First_Matches['First match \nagainst']);



# Python dictionary into a pandas DataFrame

dataFrame       = pd.DataFrame(data = inflationAndGrowth);

dataFrame.index = index;


dataFrame.plot.barh(rot=15, title="First Matches by Teams FIFA Ranks");

plot.show(block=True);
```
